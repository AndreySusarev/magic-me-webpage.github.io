## 1.5 Extra Downloads (Models, LoRAs, etc.)

```python
import subprocess
import time
import re
import sys

def run_wget_command(url, output_path, max_retries=5):
    """
    :param url: source url of the file to download
    :param output_path: path to save the downloaded files
    :param max_retries: max retry count
    """

    print("------ Starting Download ------")
    print("[url] {}".format(url))
    print("[output_path] {}".format(output_path))

    command = ["wget", "-c", url, "-O", output_path]
    try_count = 1

    while try_count <= max_retries:
            print("try {}/{}".format(try_count, max_retries))
            process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, universal_newlines=True)
            for line in iter(process.stdout.readline, ''):
                # use reg expression to find progress percentage
                match = re.search(r'(\d+)%', line)
                if match:
                    progress = int(match.group(1))
                    print(f"\rDownload Progress: {progress}%", end='')
                    sys.stdout.flush()
            return_code = process.wait()
            if return_code == 0:
                print("------ Successful Download ------")
                print("\n")
                return
            else:
                print("Failed download, retrying")
                try_count += 1
                time.sleep(try_count)

    raise RuntimeError("Failed download, reaches the MAX try times")


# Setup "source url" and "output path" of downloads
downloads_list = [
    # [source_url, output_path]
    # ["https://www.wxsnothing.com/wxs.png", "./wxs.png"],

    # Model
    ["https://civitai.com/api/download/models/130072?type=Model&format=SafeTensor&size=pruned&fp=fp16", "./models/checkpoints/realisticVision_v51.safetensors"],

    # VAE
    ["https://huggingface.co/stabilityai/sd-vae-ft-mse-original/resolve/main/vae-ft-mse-840000-ema-pruned.safetensors", "./models/vae/vae-ft-mse-840000-ema-pruned.safetensors"],

    # LoRAs
    ["https://huggingface.co/guoyww/animatediff/resolve/main/v2_lora_PanLeft.ckpt?download=true", "./custom_nodes/ComfyUI-AnimateDiff-Evolved/motion_lora/v2_lora_PanLeft.ckpt"],
    ["https://huggingface.co/guoyww/animatediff/resolve/main/v2_lora_PanRight.ckpt?download=true", "./custom_nodes/ComfyUI-AnimateDiff-Evolved/motion_lora/v2_lora_PanRight.ckpt"],
    ["https://huggingface.co/guoyww/animatediff/resolve/main/v2_lora_RollingAnticlockwise.ckpt?download=true", "./custom_nodes/ComfyUI-AnimateDiff-Evolved/motion_lora/v2_lora_RollingAnticlockwise.ckpt"],
    ["https://huggingface.co/guoyww/animatediff/resolve/main/v2_lora_RollingClockwise.ckpt?download=true", "./custom_nodes/ComfyUI-AnimateDiff-Evolved/motion_lora/v2_lora_RollingClockwise.ckpt"],
    ["https://huggingface.co/guoyww/animatediff/resolve/main/v2_lora_TiltDown.ckpt?download=true", "./custom_nodes/ComfyUI-AnimateDiff-Evolved/motion_lora/v2_lora_TiltDown.ckpt"],
    ["https://huggingface.co/guoyww/animatediff/resolve/main/v2_lora_TiltUp.ckpt?download=true", "./custom_nodes/ComfyUI-AnimateDiff-Evolved/motion_lora/v2_lora_TiltUp.ckpt"],
    ["https://huggingface.co/guoyww/animatediff/resolve/main/v2_lora_ZoomIn.ckpt?download=true", "./custom_nodes/ComfyUI-AnimateDiff-Evolved/motion_lora/v2_lora_ZoomIn.ckpt"],
    ["https://huggingface.co/guoyww/animatediff/resolve/main/v2_lora_ZoomOut.ckpt?download=true", "./custom_nodes/ComfyUI-AnimateDiff-Evolved/motion_lora/v2_lora_ZoomOut.ckpt"],

    # Animediff
    ["https://huggingface.co/guoyww/animatediff/resolve/main/mm_sd_v15_v2.ckpt?download=true", "./custom_nodes/ComfyUI-AnimateDiff-Evolved/models/mm_sd_v15_v2.ckpt"],

    # ESRGAN Upscale Model
    ["https://civitai.com/api/download/models/125843", "./models/upscale_models/4xUltrasharpV10.pt"],
]

%cd $WORKSPACE
for download_config in downloads_list:
    source_url = download_config[0]
    output_path = download_config[1]
    run_wget_command(source_url, output_path)
```

## 2.2 Run ComfyUI

```python
'''
    #############################################
    Run this cell to open ComfyUI with cloudfared
    #############################################
'''

%cd $WORKSPACE
existing_cloudflared_file_name = "cloudflared-linux-amd64.deb"
!if [ -f $existing_cloudflared_file_name ]; then rm $existing_cloudflared_file_name; fi
!wget https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
!dpkg -i ./cloudflared-linux-amd64.deb

import subprocess
import threading
import time
import socket
import urllib.request

def iframe_thread(port):
  while True:
      time.sleep(0.5)
      sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
      result = sock.connect_ex(('127.0.0.1', port))
      if result == 0:
        break
      sock.close()
  print("\nComfyUI finished loading, trying to launch cloudflared (if it gets stuck here cloudflared is having issues)\n")

  p = subprocess.Popen(["cloudflared", "tunnel", "--url", "http://127.0.0.1:{}".format(port)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
  for line in p.stderr:
    l = line.decode()
    if "trycloudflare.com " in l:
      print("This is the URL to access ComfyUI:", l[l.find("http"):], end='')
    #print(l, end='')


threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

!python main.py --dont-print-server
```
